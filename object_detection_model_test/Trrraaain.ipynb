```python
#git YOLOv5 저장소 복사
import requests
import zipfile
from io import BytesIO

url = "https://github.com/ultralytics/yolov5/archive/refs/heads/master.zip"
response = requests.get(url)

with zipfile.ZipFile(BytesIO(response.content)) as zf:
    zf.extractall()

%cd yolov5-master
```

```python
import requests

#사전에 학습된 가중치 다운로드. yolov5m.pt나 yolov5l.pt 같은 것도 있음. 자세한 건 인터넷 검색

url = "https://github.com/ultralytics/yolov5/releases/download/v5.0/yolov5s.pt"
response = requests.get(url)

with open("yolov5s.pt", "wb") as f:
    f.write(response.content)
```
