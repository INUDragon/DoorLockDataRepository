```python
#YOLOv5 저장소 복사
import requests
import zipfile
from io import BytesIO

url = "https://github.com/ultralytics/yolov5/archive/refs/heads/master.zip"
response = requests.get(url)

with zipfile.ZipFile(BytesIO(response.content)) as zf:
    zf.extractall()

%cd yolov5-master
```
